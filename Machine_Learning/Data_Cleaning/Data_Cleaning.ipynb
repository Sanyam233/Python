```python
data1 = pd.read_csv("BL-Flickr-Images-Book.csv")
data1.head()
```

```python
to_drop = ['Edition Statement','Corporate Author', 'Corporate Contributors', 'Former owner',
       'Engraver','Issuance type','Shelfmarks']
data1.drop(columns = to_drop, inplace = True)
```

```python
data1["Identifier"].is_unique
```

```python
#we normal should have a unique index which can making slicing a bit more easier 
data1 = data1.set_index("Identifier")
data1.head()
```

"""Unlike primary keys in SQL, a Pandas Index doesn’t make any guarantee of being unique, 
although many indexing and merging operations will notice a speedup in runtime if it is."""

```python
#regrex formatting was used to extract the dates from an unformatted date columns
extr = data1["Date of Publication"].str.extract(r'^(\d{4})', expand = False)
extr.head()
```

```python
#converts the dates into numeric to perform computaions on them 
data1["Date of Publication"] = pd.to_numeric(extr)
data1["Date of Publication"].dtype
```

```python
#finding the percentage of missing data 
date_precentage_missing = str(round((data1["Date of Publication"].isnull().sum() / len(data1["Date of Publication"])) * 100,2)) + " %"
print(date_precentage_missing)
```

```python
data1.head()
```

```python
data1["Place of Publication"].value_counts()
```

```python
pub1 = data1["Place of Publication"].str.contains('London')
pub2 = data1["Place of Publication"].str.contains('Oxford')
```

```python
#np.where is used to apply a condition on a series. pub1 == True if the string contains London if yes replace it with "London" else do the "Oxford" np.where statement 
data1["Place of Publication"] = np.where(pub1, "London", np.where(pub2, "Oxford",data1["Place of Publication"].str.replace("-", " ")))
```
