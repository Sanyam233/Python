```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
missing_vals = ["na", "n/a", "-", " ", "NA", "N/A", "None", "NONE", "none"]
data = pd.read_csv("dataset1.csv",na_values = missing_vals)
data.head()
```

```python
#checking what column 0 is all about
data.iloc[:,0].is_unique
#as it has all unique values there is no use of the column in our data set thus we can go head and drop it
```

```python
data = data.iloc[:,1:]
data.head()
```

```python
data.tail()
```

```python
#checking whether the data set has any missing values or no
data.isnull().sum()
```

```python
#another way of checking if any missing values exists
sns.heatmap(data.isnull(), cmap = "viridis")
```

```python
#lets see the percentage of missing data 
print("Missing Data in floor column: " +str(round((data["floor"].isnull().sum() / data["floor"].count()) * 100, 2)) + "%" )
```

```python
#filling missing values with mean
mean_value=data['floor'].mean()
data['floor']=data['floor'].fillna(mean_value)
```

```python
sns.heatmap(data.isnull(), cmap = "viridis")
```

```python
#regrex formatting was used to extract the dates from an unformatted date columns
new_list = []

for i in data.loc[:, "hoa" :]:
    extr = data[i].str.extract(r'((?:\d+,?)+)', expand = False)
    for j in extr:
        parsed = str(j).replace(",","")
        new_list.append(float(parsed))
        
    data[i] = np.array(new_list)
    new_list = []
    
    
```
